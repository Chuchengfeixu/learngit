```python
def return_list(n):
    x=[]
    y=[]
    for i in range(n):
        a=random.uniform(-1, 1)
        x.append(a)
        y.append(1/(a*a+2+2*a))
    return x,y

```

## 选取三个随机点

```python
#选取三个点作为差值基准点
x,y=return_list(3)
#numpy.concatenate(list1 , list2)
plt.plot(x,y,'or')
plt.show()
```

## 朗格朗日插值图像

```python
def Lagrange(x,y):
    #ans_x,ans_y
    ans_x=np.linspace(-1,1,100)
    ans_y=np.linspace(-1,1,100)
    for k in range(100):        
        for i in range(len(x)):
            ans_y[k]=0.0
            for j in range(len(x)):
                if i!=j:
                    y[i]*=((ans_x[k]-x[j])/(x[i]-x[j]))
                else:
                    y[i]*=1
            ans_y[k]+=y[i]
    
    plt.plot(ans_x,ans_y)
    plt.show()
Lagrange(x,y)
```
