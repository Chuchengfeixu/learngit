```python
import caffe
import numpy as np
import os
caffe.set_mode_gpu()
```

```python
if os.path.isfile('/home/zzq/result/kaggle/result.csv'):
    os.remove('/home/zzq/result/kaggle/result.csv')
net=caffe.Net('/home/zzq/caffe/models/bvlc_googlenet/deploy.prototxt',
             '/home/zzq/caffe/models/bvlc_googlenet/bvlc_googlenet_iter_1200.caffemodel',
             caffe.TEST)
lines1=open('/home/zzq/result/kaggle/image_list','r').readlines()
root1='/home/zzq/kaggle/fish/test_stg1/'
label='/home/zzq/result/kaggle/label'
la=open(label,'r').readlines()
f=file("/home/zzq/result/kaggle/result.csv","a+")
str_='image,'
for i in range(0,len(la)):
    l_=la[i].strip('\n')
    if i==len(la)-1:
        str_=str_+l_
    else:
        str_=str_+l_+','
f.writelines(str_+'\n')
for i in range(0,len(lines1)):    
    #transformer.set_raw_scale('data', 255)
    lines_=lines1[i].strip('\n')
    img=root1+lines_
    transformer = caffe.io.Transformer({'data': net.blobs['data'].data.shape})  #设定图片的shape格式(1,3,28,28)
    transformer.set_transpose('data', (2,0,1))    #改变维度的顺序，由原始图片(28,28,3)变为(3,28,28)
    #transformer.set_mean('data', np.load(mean_file).mean(1).mean(1))    #减去均值，前面训练模型时没有减均值，这儿就不用
    transformer.set_raw_scale('data', 255)    # 缩放到【0，255】之间
    transformer.set_channel_swap('data', (2,1,0))   #交换通道，将图片由RGB变为BGR
    im=caffe.io.load_image(img) 
    #net.blobs['data'].data[...] = transformer.preprocess('data',im) 
    #print lines[i]
    #net.blobs['data']=im
    net.blobs['data'].data[...] = transformer.preprocess('data',im) 
    out=net.forward()
    prob=net.blobs['prob'].data[0].flatten()
    f=file("/home/zzq/result/kaggle/result.csv","a+")
    xian=lines_+','+('%.10f,%.10f,%.10f,%.10f,%.10f,%.10f,%.10f,%.10f'%(prob[0],
                                                                         prob[1],
                                                                         prob[2],
                                                                         prob[3],
                                                                         prob[4],
                                                                         prob[5],
                                                                          prob[6],
                                                                          prob[7]
                                                                         )) + '\n'
    f.writelines(xian)
#     if i==len(lines1)-1:
#         img_1001='/home/zzq/kaggle/test_crop/img_07921.jpg'
#         im_1001=caffe.io.load_image(img) 
#         net.blobs['data'].data[...] = transformer.preprocess('data',im_1001)
#         out=net.forward()
#         prob=net.blobs['prob'].data[0].flatten()
#         f=file("/home/zzq/result/kaggle/result.csv","a+")
#         xian_='img_07921.jpg'+','+('%.10f,%.10f,%.10f,%.10f,%.10f,%.10f,%.10f,%.10f'%(prob[0],
#                                                                          prob[1],
#                                                                          prob[2],
#                                                                          prob[3],
#                                                                          prob[4],
#                                                                          prob[5],
#                                                                           prob[6],
#                                                                           prob[7]
#                                                                          )) + '\n'
#         f.writelines(xian_)
        
lines2=open('/home/zzq/result/kaggle/image_stg2','r').readlines()
root2='/home/zzq/kaggle/fish/test_stg2/'
for i in range(0,len(lines2)):
    #transformer.set_raw_scale('data', 255)
    lines_=lines2[i].strip('\n')#去除换行
    img=root2+lines_
    transformer = caffe.io.Transformer({'data': net.blobs['data'].data.shape})  #设定图片的shape格式(1,3,28,28)
    transformer.set_transpose('data', (2,0,1))    #改变维度的顺序，由原始图片(28,28,3)变为(3,28,28)
    #transformer.set_mean('data', np.load(mean_file).mean(1).mean(1))    #减去均值，前面训练模型时没有减均值，这儿就不用
    transformer.set_raw_scale('data', 255)    # 缩放到【0，255】之间
    transformer.set_channel_swap('data', (2,1,0))   #交换通道，将图片由RGB变为BGR
    im=caffe.io.load_image(img) 
    #net.blobs['data'].data[...] = transformer.preprocess('data',im) 
    #print lines[i]
    #net.blobs['data']=im
    net.blobs['data'].data[...] = transformer.preprocess('data',im) 
    out=net.forward()
    prob=net.blobs['prob'].data[0].flatten() 
    f=file("/home/zzq/result/kaggle/result.csv","a+")
    xian='test_stg2/'+lines_+','+('%.10f,%.10f,%.10f,%.10f,%.10f,%.10f,%.10f,%.10f'%(prob[0],
                                                                         prob[1],
                                                                         prob[2],
                                                                         prob[3],
                                                                         prob[4],
                                                                         prob[5],
                                                                          prob[6],
                                                                          prob[7]
                                                                         )) + '\n'
    f.writelines(xian)
    if i==len(lines2)-1:
        img_1001='/home/zzq/kaggle/fish/test_stg2/image_12153.jpg'
        im_1001=caffe.io.load_image(img) 
        net.blobs['data'].data[...] = transformer.preprocess('data',im_1001)
        out=net.forward()
        prob=net.blobs['prob'].data[0].flatten()
        f=file("/home/zzq/result/kaggle/result.csv","a+")
        xian_='test_stg2/image_12153.jpg'+','+('%.10f,%.10f,%.10f,%.10f,%10f,%.10f,%.10f,%.10f'%(prob[0],
                                                                         prob[1],
                                                                         prob[2],
                                                                         prob[3],
                                                                         prob[4],
                                                                         prob[5],
                                                                          prob[6],
                                                                          prob[7]
                                                                         )) + '\n'
        f.writelines(xian_)
```
